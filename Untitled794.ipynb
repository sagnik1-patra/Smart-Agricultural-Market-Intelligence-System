```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# =========================
# PATH SETTINGS
# =========================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# LOAD DATA
# =========================

df = pd.read_excel(DATASET_PATH)

print("Dataset Loaded Successfully")
print(df.head())
print(df.columns)

# =========================
# CLEAN COLUMN NAMES
# =========================

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("Cleaned Columns:", df.columns.tolist())

# Expected columns:
# district_name, block_name, market_name

# =========================
# BASIC CLEANING
# =========================

df = df.dropna()
df = df.drop_duplicates()

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

# =========================
# FEATURE ENGINEERING
# =========================

district_market_count = df.groupby("district_name")["market_name"].count().to_dict()
block_market_count = df.groupby("block_name")["market_name"].count().to_dict()

df["district_market_count"] = df["district_name"].map(district_market_count)
df["block_market_count"] = df["block_name"].map(block_market_count)

df["market_name_length"] = df["market_name"].apply(len)
df["block_name_length"] = df["block_name"].apply(len)
df["district_name_length"] = df["district_name"].apply(len)

# Demand / accessibility score
df["market_score"] = (
    df["district_market_count"] * 0.4 +
    df["block_market_count"] * 0.5 +
    df["market_name_length"] * 0.1
)

# Create target category
df["market_category"] = pd.qcut(
    df["market_score"],
    q=3,
    labels=["Low", "Medium", "High"],
    duplicates="drop"
)

# =========================
# ENCODING
# =========================

label_encoders = {}

for col in ["district_name", "block_name", "market_name", "market_category"]:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

# =========================
# FEATURES AND TARGET
# =========================

features = [
    "district_name_encoded",
    "block_name_encoded",
    "market_name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "market_score"
]

X = df[features]
y = df["market_category_encoded"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# CLUSTERING
# =========================

kmeans = KMeans(n_clusters=3, random_state=42)
df["market_cluster"] = kmeans.fit_predict(X_scaled)

# =========================
# TRAIN TEST SPLIT
# =========================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y if len(np.unique(y)) > 1 else None
)

# =========================
# RANDOM FOREST MODEL
# =========================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=10
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

print("Random Forest Accuracy:", rf_accuracy)

# =========================
# NEURAL NETWORK MODEL
# =========================

num_classes = len(np.unique(y))
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

nn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = nn_model.fit(
    X_train,
    y_train_cat,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test_cat),
    verbose=1
)

nn_loss, nn_accuracy = nn_model.evaluate(X_test, y_test_cat, verbose=0)

print("Neural Network Accuracy:", nn_accuracy)

# =========================
# SAVE FILES
# =========================

# Save H5 model
h5_path = os.path.join(OUTPUT_DIR, "market_intelligence_model.h5")
nn_model.save(h5_path)

# Save PKL model
pkl_path = os.path.join(OUTPUT_DIR, "market_intelligence_rf_model.pkl")
joblib.dump(rf_model, pkl_path)

# Save scaler
scaler_path = os.path.join(OUTPUT_DIR, "market_scaler.pkl")
joblib.dump(scaler, scaler_path)

# Save encoders
encoder_path = os.path.join(OUTPUT_DIR, "market_label_encoders.pkl")
joblib.dump(label_encoders, encoder_path)

# Save KMeans model
kmeans_path = os.path.join(OUTPUT_DIR, "market_kmeans_model.pkl")
joblib.dump(kmeans, kmeans_path)

# Save YAML config
config = {
    "project_name": "Smart Agricultural Market Intelligence System",
    "author": "Sagnik Patra",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "features_used": features,
    "target": "market_category",
    "models": {
        "random_forest": "market_intelligence_rf_model.pkl",
        "neural_network": "market_intelligence_model.h5",
        "kmeans": "market_kmeans_model.pkl"
    }
}

yaml_path = os.path.join(OUTPUT_DIR, "market_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, default_flow_style=False)

# Save JSON metrics
metrics = {
    "random_forest_accuracy": float(rf_accuracy),
    "neural_network_accuracy": float(nn_accuracy),
    "total_records": int(len(df)),
    "total_districts": int(df["district_name"].nunique()),
    "total_blocks": int(df["block_name"].nunique()),
    "total_markets": int(df["market_name"].nunique())
}

json_path = os.path.join(OUTPUT_DIR, "market_metrics.json")

with open(json_path, "w") as f:
    json.dump(metrics, f, indent=4)

# Save processed CSV
processed_csv_path = os.path.join(OUTPUT_DIR, "market_processed_data.csv")
df.to_csv(processed_csv_path, index=False)

# Save prediction CSV
prediction_df = df.copy()
prediction_df["predicted_market_category_encoded"] = rf_model.predict(X_scaled)

category_encoder = label_encoders["market_category"]
prediction_df["predicted_market_category"] = category_encoder.inverse_transform(
    prediction_df["predicted_market_category_encoded"]
)

prediction_csv_path = os.path.join(OUTPUT_DIR, "market_predictions.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

# =========================
# FINAL OUTPUT
# =========================

print("\nAll files saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated Files:")
print("1. market_intelligence_model.h5")
print("2. market_intelligence_rf_model.pkl")
print("3. market_scaler.pkl")
print("4. market_label_encoders.pkl")
print("5. market_kmeans_model.pkl")
print("6. market_config.yaml")
print("7. market_metrics.json")
print("8. market_processed_data.csv")
print("9. market_predictions.csv")
```
