```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.cluster import KMeans

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# =========================
# PATH SETTINGS
# =========================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "csa_"

# =========================
# LOAD DATA
# =========================

df = pd.read_excel(DATASET_PATH)

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df = df.dropna()
df = df.drop_duplicates()

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

print("Dataset Loaded Successfully")
print(df.head())

# =========================
# FEATURE ENGINEERING
# =========================

district_market_count = df.groupby("district_name")["market_name"].count().to_dict()
block_market_count = df.groupby("block_name")["market_name"].count().to_dict()

df["district_market_count"] = df["district_name"].map(district_market_count)
df["block_market_count"] = df["block_name"].map(block_market_count)

df["market_name_length"] = df["market_name"].apply(len)
df["block_name_length"] = df["block_name"].apply(len)
df["district_name_length"] = df["district_name"].apply(len)

df["market_score"] = (
    df["district_market_count"] * 0.4 +
    df["block_market_count"] * 0.5 +
    df["market_name_length"] * 0.1
)

df["market_category"] = pd.qcut(
    df["market_score"],
    q=3,
    labels=["Low", "Medium", "High"],
    duplicates="drop"
)

# =========================
# ENCODING
# =========================

label_encoders = {}

for col in ["district_name", "block_name", "market_name", "market_category"]:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

# =========================
# FEATURES AND TARGET
# =========================

features = [
    "district_name_encoded",
    "block_name_encoded",
    "market_name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "market_score"
]

X = df[features]
y = df["market_category_encoded"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# CSA FEATURE SELECTION
# Clonal Selection Algorithm
# =========================

def fitness_function(mask, X_data, y_data):
    selected_features = np.where(mask == 1)[0]

    if len(selected_features) == 0:
        return 0

    X_selected = X_data[:, selected_features]

    X_train, X_test, y_train, y_test = train_test_split(
        X_selected,
        y_data,
        test_size=0.2,
        random_state=42,
        stratify=y_data if len(np.unique(y_data)) > 1 else None
    )

    model = RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        max_depth=8
    )

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    accuracy = accuracy_score(y_test, preds)
    feature_penalty = len(selected_features) / X_data.shape[1]

    fitness = accuracy - (0.01 * feature_penalty)

    return fitness


def csa_feature_selection(
    X_data,
    y_data,
    population_size=25,
    generations=30,
    clone_rate=5,
    mutation_rate=0.15,
    random_cells=5
):
    n_features = X_data.shape[1]

    population = np.random.randint(0, 2, size=(population_size, n_features))

    for i in range(population_size):
        if np.sum(population[i]) == 0:
            population[i][np.random.randint(0, n_features)] = 1

    best_solution = None
    best_fitness = -np.inf
    fitness_history = []

    for generation in range(generations):
        fitness_scores = np.array([
            fitness_function(individual, X_data, y_data)
            for individual in population
        ])

        sorted_indices = np.argsort(fitness_scores)[::-1]
        population = population[sorted_indices]
        fitness_scores = fitness_scores[sorted_indices]

        if fitness_scores[0] > best_fitness:
            best_fitness = fitness_scores[0]
            best_solution = population[0].copy()

        selected_population = population[:population_size // 2]

        clones = []

        for rank, antibody in enumerate(selected_population):
            number_of_clones = clone_rate * (len(selected_population) - rank)

            for _ in range(number_of_clones):
                clone = antibody.copy()

                adaptive_mutation_rate = mutation_rate * (rank + 1) / len(selected_population)

                for j in range(n_features):
                    if np.random.rand() < adaptive_mutation_rate:
                        clone[j] = 1 - clone[j]

                if np.sum(clone) == 0:
                    clone[np.random.randint(0, n_features)] = 1

                clones.append(clone)

        clones = np.array(clones)

        combined_population = np.vstack((population, clones))

        combined_fitness = np.array([
            fitness_function(individual, X_data, y_data)
            for individual in combined_population
        ])

        best_indices = np.argsort(combined_fitness)[::-1][:population_size - random_cells]

        best_cells = combined_population[best_indices]

        new_random_cells = np.random.randint(0, 2, size=(random_cells, n_features))

        for i in range(random_cells):
            if np.sum(new_random_cells[i]) == 0:
                new_random_cells[i][np.random.randint(0, n_features)] = 1

        population = np.vstack((best_cells, new_random_cells))

        fitness_history.append(best_fitness)

        print(
            f"Generation {generation + 1}/{generations} | "
            f"Best Fitness: {best_fitness:.4f} | "
            f"Selected Features: {np.sum(best_solution)}"
        )

    return best_solution, best_fitness, fitness_history


best_mask, best_fitness, fitness_history = csa_feature_selection(X_scaled, y)

selected_feature_indices = np.where(best_mask == 1)[0]
selected_features = [features[i] for i in selected_feature_indices]

print("\nSelected Features by CSA:")
print(selected_features)

X_selected = X_scaled[:, selected_feature_indices]

# =========================
# CLUSTERING
# =========================

kmeans = KMeans(n_clusters=3, random_state=42)
df["csa_market_cluster"] = kmeans.fit_predict(X_selected)

# =========================
# TRAIN TEST SPLIT
# =========================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y if len(np.unique(y)) > 1 else None
)

# =========================
# RANDOM FOREST MODEL
# =========================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=10
)

rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

# =========================
# GRADIENT BOOSTING MODEL
# =========================

gb_model = GradientBoostingClassifier(random_state=42)

gb_model.fit(X_train, y_train)
gb_pred = gb_model.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_pred)

# =========================
# NEURAL NETWORK MODEL
# =========================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

nn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = nn_model.fit(
    X_train,
    y_train_cat,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test_cat),
    verbose=1
)

nn_loss, nn_accuracy = nn_model.evaluate(X_test, y_test_cat, verbose=0)

# =========================
# SAVE MODELS
# =========================

nn_model.save(os.path.join(OUTPUT_DIR, PREFIX + "market_intelligence_model.h5"))

joblib.dump(rf_model, os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl"))
joblib.dump(gb_model, os.path.join(OUTPUT_DIR, PREFIX + "gradient_boosting_model.pkl"))
joblib.dump(kmeans, os.path.join(OUTPUT_DIR, PREFIX + "kmeans_model.pkl"))
joblib.dump(scaler, os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl"))
joblib.dump(label_encoders, os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl"))
joblib.dump(best_mask, os.path.join(OUTPUT_DIR, PREFIX + "selected_feature_mask.pkl"))

# =========================
# RESULT CSV
# =========================

result_df = pd.DataFrame({
    "Model": [
        "CSA + Random Forest",
        "CSA + Gradient Boosting",
        "CSA + Neural Network"
    ],
    "Accuracy": [
        rf_accuracy,
        gb_accuracy,
        nn_accuracy
    ]
})

result_df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "result.csv"), index=False)

# =========================
# PREDICTION CSV
# =========================

df["csa_rf_prediction_encoded"] = rf_model.predict(X_selected)
df["csa_gb_prediction_encoded"] = gb_model.predict(X_selected)

nn_full_pred = nn_model.predict(X_selected)
df["csa_nn_prediction_encoded"] = np.argmax(nn_full_pred, axis=1)

category_encoder = label_encoders["market_category"]

df["csa_rf_prediction"] = category_encoder.inverse_transform(
    df["csa_rf_prediction_encoded"]
)

df["csa_gb_prediction"] = category_encoder.inverse_transform(
    df["csa_gb_prediction_encoded"]
)

df["csa_nn_prediction"] = category_encoder.inverse_transform(
    df["csa_nn_prediction_encoded"]
)

df.to_csv(os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv"), index=False)

# =========================
# JSON METRICS
# =========================

metrics = {
    "optimization_algorithm": "Clonal Selection Algorithm",
    "csa_best_fitness": float(best_fitness),
    "selected_features": selected_features,
    "random_forest_accuracy": float(rf_accuracy),
    "gradient_boosting_accuracy": float(gb_accuracy),
    "neural_network_accuracy": float(nn_accuracy),
    "total_records": int(len(df)),
    "total_districts": int(df["district_name"].nunique()),
    "total_blocks": int(df["block_name"].nunique()),
    "total_markets": int(df["market_name"].nunique())
}

with open(os.path.join(OUTPUT_DIR, PREFIX + "metrics.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# =========================
# YAML CONFIG
# =========================

config = {
    "project_name": "Smart Agricultural Market Intelligence System",
    "author": "Sagnik Patra",
    "optimization_algorithm": "Clonal Selection Algorithm",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "all_features": features,
    "selected_features": selected_features,
    "target": "market_category",
    "models": {
        "random_forest": PREFIX + "random_forest_model.pkl",
        "gradient_boosting": PREFIX + "gradient_boosting_model.pkl",
        "neural_network": PREFIX + "market_intelligence_model.h5",
        "kmeans": PREFIX + "kmeans_model.pkl"
    },
    "outputs": {
        "result_csv": PREFIX + "result.csv",
        "prediction_csv": PREFIX + "prediction.csv",
        "accuracy_graph": PREFIX + "accuracy_graph.png",
        "heatmap": PREFIX + "heatmap.png",
        "comparison_graph": PREFIX + "comparison_graph.png",
        "result_graph": PREFIX + "result_graph.png",
        "prediction_graph": PREFIX + "prediction_graph.png",
        "optimization_graph": PREFIX + "optimization_graph.png",
        "confusion_matrix": PREFIX + "confusion_matrix.png"
    }
}

with open(os.path.join(OUTPUT_DIR, PREFIX + "config.yaml"), "w") as f:
    yaml.dump(config, f, default_flow_style=False)

# =========================
# GRAPH 1: ACCURACY GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("CSA Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png"))
plt.show()

# =========================
# GRAPH 2: HEATMAP
# =========================

plt.figure(figsize=(10, 7))
heatmap_data = df[selected_features + ["market_category_encoded", "csa_market_cluster"]]
sns.heatmap(heatmap_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("CSA Selected Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png"))
plt.show()

# =========================
# GRAPH 3: COMPARISON GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.plot(result_df["Model"], result_df["Accuracy"], marker="o")
plt.title("CSA Model Performance Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(True)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png"))
plt.show()

# =========================
# GRAPH 4: RESULT GRAPH
# =========================

plt.figure(figsize=(8, 5))
actual_counts = df["market_category"].value_counts()
plt.bar(actual_counts.index, actual_counts.values)
plt.title("Actual Market Category Distribution")
plt.xlabel("Market Category")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png"))
plt.show()

# =========================
# GRAPH 5: PREDICTION GRAPH
# =========================

plt.figure(figsize=(8, 5))
pred_counts = df["csa_rf_prediction"].value_counts()
plt.bar(pred_counts.index, pred_counts.values)
plt.title("CSA Predicted Market Category Distribution")
plt.xlabel("Predicted Category")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png"))
plt.show()

# =========================
# GRAPH 6: CSA OPTIMIZATION GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(fitness_history) + 1), fitness_history, marker="o")
plt.title("CSA Optimization Progress")
plt.xlabel("Generation")
plt.ylabel("Best Fitness")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "optimization_graph.png"))
plt.show()

# =========================
# GRAPH 7: NEURAL NETWORK TRAINING ACCURACY
# =========================

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("CSA Neural Network Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "nn_training_accuracy_graph.png"))
plt.show()

# =========================
# GRAPH 8: CONFUSION MATRIX
# =========================

cm = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("CSA Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "confusion_matrix.png"))
plt.show()

# =========================
# FINAL OUTPUT
# =========================

print("\nAll CSA files saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated CSA Files:")
print(PREFIX + "market_intelligence_model.h5")
print(PREFIX + "random_forest_model.pkl")
print(PREFIX + "gradient_boosting_model.pkl")
print(PREFIX + "kmeans_model.pkl")
print(PREFIX + "scaler.pkl")
print(PREFIX + "label_encoders.pkl")
print(PREFIX + "selected_feature_mask.pkl")
print(PREFIX + "config.yaml")
print(PREFIX + "metrics.json")
print(PREFIX + "result.csv")
print(PREFIX + "prediction.csv")
print(PREFIX + "accuracy_graph.png")
print(PREFIX + "heatmap.png")
print(PREFIX + "comparison_graph.png")
print(PREFIX + "result_graph.png")
print(PREFIX + "prediction_graph.png")
print(PREFIX + "optimization_graph.png")
print(PREFIX + "nn_training_accuracy_graph.png")
print(PREFIX + "confusion_matrix.png")
```
