```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# =========================
# PATH SETTINGS
# =========================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Market Intelligence System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# LOAD DATA
# =========================

df = pd.read_excel(DATASET_PATH)

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df = df.dropna()
df = df.drop_duplicates()

for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

print("Dataset Loaded Successfully")
print(df.head())

# =========================
# FEATURE ENGINEERING
# =========================

district_market_count = df.groupby("district_name")["market_name"].count().to_dict()
block_market_count = df.groupby("block_name")["market_name"].count().to_dict()

df["district_market_count"] = df["district_name"].map(district_market_count)
df["block_market_count"] = df["block_name"].map(block_market_count)

df["market_name_length"] = df["market_name"].apply(len)
df["block_name_length"] = df["block_name"].apply(len)
df["district_name_length"] = df["district_name"].apply(len)

df["market_score"] = (
    df["district_market_count"] * 0.4 +
    df["block_market_count"] * 0.5 +
    df["market_name_length"] * 0.1
)

df["market_category"] = pd.qcut(
    df["market_score"],
    q=3,
    labels=["Low", "Medium", "High"],
    duplicates="drop"
)

# =========================
# ENCODING
# =========================

label_encoders = {}

for col in ["district_name", "block_name", "market_name", "market_category"]:
    le = LabelEncoder()
    df[col + "_encoded"] = le.fit_transform(df[col])
    label_encoders[col] = le

# =========================
# FEATURES AND TARGET
# =========================

features = [
    "district_name_encoded",
    "block_name_encoded",
    "market_name_encoded",
    "district_market_count",
    "block_market_count",
    "market_name_length",
    "block_name_length",
    "district_name_length",
    "market_score"
]

X = df[features]
y = df["market_category_encoded"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =========================
# CLUSTERING
# =========================

kmeans = KMeans(n_clusters=3, random_state=42)
df["market_cluster"] = kmeans.fit_predict(X_scaled)

# =========================
# TRAIN TEST SPLIT
# =========================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y if len(np.unique(y)) > 1 else None
)

# =========================
# RANDOM FOREST MODEL
# =========================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=10
)

rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_pred)

# =========================
# GRADIENT BOOSTING MODEL
# =========================

gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)
gb_pred = gb_model.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_pred)

# =========================
# NEURAL NETWORK MODEL
# =========================

num_classes = len(np.unique(y))

y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat = to_categorical(y_test, num_classes=num_classes)

nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(num_classes, activation="softmax")
])

nn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = nn_model.fit(
    X_train,
    y_train_cat,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test_cat),
    verbose=1
)

nn_loss, nn_accuracy = nn_model.evaluate(X_test, y_test_cat, verbose=0)

# =========================
# SAVE MODELS
# =========================

nn_model.save(os.path.join(OUTPUT_DIR, "market_intelligence_model.h5"))

joblib.dump(rf_model, os.path.join(OUTPUT_DIR, "market_random_forest_model.pkl"))
joblib.dump(gb_model, os.path.join(OUTPUT_DIR, "market_gradient_boosting_model.pkl"))
joblib.dump(kmeans, os.path.join(OUTPUT_DIR, "market_kmeans_model.pkl"))
joblib.dump(scaler, os.path.join(OUTPUT_DIR, "market_scaler.pkl"))
joblib.dump(label_encoders, os.path.join(OUTPUT_DIR, "market_label_encoders.pkl"))

# =========================
# RESULT CSV
# =========================

result_df = pd.DataFrame({
    "Model": ["Random Forest", "Gradient Boosting", "Neural Network"],
    "Accuracy": [rf_accuracy, gb_accuracy, nn_accuracy]
})

result_csv_path = os.path.join(OUTPUT_DIR, "market_result.csv")
result_df.to_csv(result_csv_path, index=False)

# =========================
# PREDICTION CSV
# =========================

df["rf_prediction_encoded"] = rf_model.predict(X_scaled)
df["gb_prediction_encoded"] = gb_model.predict(X_scaled)

nn_full_pred = nn_model.predict(X_scaled)
df["nn_prediction_encoded"] = np.argmax(nn_full_pred, axis=1)

category_encoder = label_encoders["market_category"]

df["rf_prediction"] = category_encoder.inverse_transform(df["rf_prediction_encoded"])
df["gb_prediction"] = category_encoder.inverse_transform(df["gb_prediction_encoded"])
df["nn_prediction"] = category_encoder.inverse_transform(df["nn_prediction_encoded"])

prediction_csv_path = os.path.join(OUTPUT_DIR, "market_prediction.csv")
df.to_csv(prediction_csv_path, index=False)

# =========================
# JSON METRICS
# =========================

metrics = {
    "random_forest_accuracy": float(rf_accuracy),
    "gradient_boosting_accuracy": float(gb_accuracy),
    "neural_network_accuracy": float(nn_accuracy),
    "total_records": int(len(df)),
    "total_districts": int(df["district_name"].nunique()),
    "total_blocks": int(df["block_name"].nunique()),
    "total_markets": int(df["market_name"].nunique())
}

with open(os.path.join(OUTPUT_DIR, "market_metrics.json"), "w") as f:
    json.dump(metrics, f, indent=4)

# =========================
# YAML CONFIG
# =========================

config = {
    "project_name": "Smart Agricultural Market Intelligence System",
    "author": "Sagnik Patra",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "features_used": features,
    "target": "market_category",
    "models": {
        "random_forest": "market_random_forest_model.pkl",
        "gradient_boosting": "market_gradient_boosting_model.pkl",
        "neural_network": "market_intelligence_model.h5",
        "kmeans": "market_kmeans_model.pkl"
    },
    "outputs": {
        "result_csv": "market_result.csv",
        "prediction_csv": "market_prediction.csv",
        "accuracy_graph": "market_accuracy_graph.png",
        "heatmap": "market_heatmap.png",
        "comparison_graph": "market_comparison_graph.png",
        "result_graph": "market_result_graph.png",
        "prediction_graph": "market_prediction_graph.png"
    }
}

with open(os.path.join(OUTPUT_DIR, "market_config.yaml"), "w") as f:
    yaml.dump(config, f, default_flow_style=False)

# =========================
# GRAPH 1: ACCURACY GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.bar(result_df["Model"], result_df["Accuracy"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_accuracy_graph.png"))
plt.show()

# =========================
# GRAPH 2: HEATMAP
# =========================

plt.figure(figsize=(10, 7))
corr_data = df[features + ["market_category_encoded", "market_cluster"]]
sns.heatmap(corr_data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_heatmap.png"))
plt.show()

# =========================
# GRAPH 3: COMPARISON GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.plot(result_df["Model"], result_df["Accuracy"], marker="o")
plt.title("Model Performance Comparison")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_comparison_graph.png"))
plt.show()

# =========================
# GRAPH 4: RESULT GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.bar(df["market_category"].value_counts().index, df["market_category"].value_counts().values)
plt.title("Actual Market Category Distribution")
plt.xlabel("Market Category")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_result_graph.png"))
plt.show()

# =========================
# GRAPH 5: PREDICTION GRAPH
# =========================

plt.figure(figsize=(8, 5))
plt.bar(df["rf_prediction"].value_counts().index, df["rf_prediction"].value_counts().values)
plt.title("Predicted Market Category Distribution - Random Forest")
plt.xlabel("Predicted Category")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_prediction_graph.png"))
plt.show()

# =========================
# GRAPH 6: NEURAL NETWORK TRAINING ACCURACY
# =========================

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("Neural Network Training Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_nn_training_accuracy_graph.png"))
plt.show()

# =========================
# GRAPH 7: CONFUSION MATRIX
# =========================

cm = confusion_matrix(y_test, rf_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "market_confusion_matrix.png"))
plt.show()

# =========================
# FINAL PRINT
# =========================

print("\nAll files saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated Files:")
print("1. market_intelligence_model.h5")
print("2. market_random_forest_model.pkl")
print("3. market_gradient_boosting_model.pkl")
print("4. market_kmeans_model.pkl")
print("5. market_scaler.pkl")
print("6. market_label_encoders.pkl")
print("7. market_config.yaml")
print("8. market_metrics.json")
print("9. market_result.csv")
print("10. market_prediction.csv")
print("11. market_accuracy_graph.png")
print("12. market_heatmap.png")
print("13. market_comparison_graph.png")
print("14. market_result_graph.png")
print("15. market_prediction_graph.png")
print("16. market_nn_training_accuracy_graph.png")
print("17. market_confusion_matrix.png")
```
